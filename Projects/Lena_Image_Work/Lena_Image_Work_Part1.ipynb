### Resizing

RGB is the format what we are looking for. On the other hand size of image is 512*512. However we need 256*256. So we need to convert it.

```python
im_rs=im.resize((256,256));im_rs.size
```

## Gray Scaling

```python
im_arr= np.array(im_rs)
```

```python
a=len(im_arr)
b=len(im_arr)

for i in range(0,a):
    
    for j in range(0,b):
        
        im_arr[i,j][0]=im_arr[i,j].mean()
```
