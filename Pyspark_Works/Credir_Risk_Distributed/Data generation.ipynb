```python
import pandas as pd
import numpy as np
from imblearn.over_sampling import RandomOverSampler
```

```python
df=pd.read_csv("data.csv",low_memory=False)
```

```python
df.dropna(axis=1,inplace=True)
df.drop(["Unnamed: 0","ID"],axis=1,inplace=True)
```
